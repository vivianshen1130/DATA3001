```python
#Reading data_1.txt adding an index so I can read it with pd.read_json
lists = {}
with open("data_1.txt","r") as file:
    data = json.load(file)
    num = 0
    for i in data:
        for j in data[i].keys():
            num += 1
            player = j
            lists[num] = data[i][j]
            lists[num]['player'] = player
        
```

```python
#write list to data_2.txt
with open('data_2.txt', 'w') as outfile:
    json.dump(lists, outfile)
```
